```python
import numpy as np

import matplotlib.pyplot as pl 
% matplotlib inline
pl.rcParams['figure.figsize'] = ( 12.0, 6.0 )

from mpltools import style
from mpltools import special
style.use( 'ggplot' )

import GPy 
```

```python
"""
Some convenience visualisation functions
"""

def plotKernel( kernel, x ):
    fig = pl.figure( figsize = ( 5, 5 ) )
    xx, yy = np.meshgrid( x, x )
    pl.pcolor( xx, yy, kernel.K( np.c_[x] ) )
    pl.colorbar()
    
def test( model, x ): 
    x_test = np.linspace( x.min() * 5, x.max() * 5, N * 5 )
    mean, variance = model.predict( np.c_[x_test] )
    print mean.shape, variance.shape
    special.errorfill( x_test, mean[:,0], 2 * np.sqrt( variance[:,0] ) )
```

```python
"""
Set up the data for this test as consisting of two periodic signals with the addition of noise
"""

N = 100 
x = np.linspace( -10, 10, N ) 
y = np.sin( x * 3.0 ) + np.sin( x / 3.0 )
n = np.random.randn( N ) * 0.5

pl.plot( x, y, x, y + n )
```

```python
"""
Test the model when no noise is introduced
"""

kernel = GPy.kern.RBF( 1 ) 
model_clean = GPy.models.GPRegression( np.c_[x], np.c_[y], kernel )

print model_clean
```

```python
"""
Show the performance of the model with the default parameters
"""

model_clean.plot()
```

```python
"""
And lets look at the covariance kernel
Because our inputs are ordered according to time (our only feature) the covariance matrix appears as a diagonal streak. 
"""

plotKernel( kernel, x ) 

print kernel

kernel.lengthscale *= 2
plotKernel( kernel, x )
```

```python
"""
Use the in-build optimisation of the classification model in order to select the best hyper parameters 
(I believe this uses a conjugate gradient optimiser.)
"""

model_clean.randomize()
model_clean.optimize()
model_clean.plot()

print model_clean
```

```python
"""
Look at the optimised covariance matrix - it hasn't changed too much
"""

plotKernel( kernel, x ) 
```

```python
"""
See how the regression looks outside of the training range
"""
    
test( model_clean, x ) 
pl.plot( x, y ) 
```

```python
"""
Repeat the above with noisy observations
"""

kernel = GPy.kern.RBF( 1 )
model = GPy.models.GPRegression( np.c_[x], np.c_[y + n], kernel )

print model 
```

```python
"""
Optimise the model and print its parameters
"""

model.optimize()
model.plot()

print model 
```

```python
"""
See how this model works outside of the training range
"""

test( model, x ) 
pl.plot( x, y )
```

```python
"""
Lets assume that we know our data consists of the superposition of two periodic variables with noise. 
Lets introduce this knowledge in the form of a kernel function
"""

kernel = GPy.kern.Cosine( 1 ) + GPy.kern.Cosine( 1 ) 

print kernel 
```

```python
"""
Now we introduce the periodic kernel to the GP model 
"""

model = GPy.models.GPRegression( np.c_[x], np.c_[y + n], kernel )
model.randomize()
model.optimize()
model.plot()

print model 
```

```python
"""
Even with the introduction of periodic kernels, the generalisation doesn't look too great
outside of the training range
"""

test( model, x ) 
pl.plot( x, y ) 
```

```python
"""
However, recall, that the hyperparameter space is non-convex
"""

model.optimize_restarts( 10 )
model.plot()

print model 
```

```python
"""
And this generalises much better outside the input range (in blue)
"""

test( model, x ) 
pl.plot( x, y, x, y + n )
```

```python
""" 
And now lets plot the kernel - notice a high- and low-frequency contingent to the image
"""

plotKernel( kernel, x ) 
print kernel
```

```python
"""
Now, we will introduce constraints. 
Lets assume that we know the fast sinusoid is between 0.2 and 0.4 Hz while the slow one is between 2 and 4 Hz
"""

periodicKernel_1 = GPy.kern.Cosine( 1 )
periodicKernel_2 = GPy.kern.Cosine( 1 )

kernel = periodicKernel_1 + periodicKernel_2

print kernel 

periodicKernel_1.lengthscale.constrain_bounded( 0.2, 0.4 )
periodicKernel_2.lengthscale.constrain_bounded( 2.0, 4.0 )

print kernel
```

```python
"""
Test the constrained kernel. Note that the lengthscale of both cosine components 
corresponds to the multiplicative factor set in the initial data (approximately 1/3 and 3)
"""

model = GPy.models.GPRegression( np.c_[x], np.c_[y + n], kernel )

model.randomize()
model.optimize_restarts( 10 )
model.plot()

print model 
```

```python
test( model, x ) 
pl.plot( x, y )
```
